```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from statsmodels.graphics import tsaplots
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from IPython.display import display, HTML

%matplotlib inline 
# Set default style
sns.set()
```

```python
%load_ext watermark
%watermark -iv -u -n -t -z
```

```python
# File path
outbounds_file = 'Dataset/outbounds14to17.csv'
```

```python
# Read in data as DataFrame
outbounds = pd.read_csv(outbounds_file, index_col=0, parse_dates=True)
display(outbounds)
```

```python
# Overall info
display(outbounds.info())
display(outbounds.describe())
```

```python
# Autocorrelation
fig = tsaplots.plot_acf(outbounds['Grand Total'],lags=12)
fig2 = tsaplots.plot_pacf(outbounds['Grand Total'], lags=12)
plt.show()
```

```python
# Create list for regions subset
regions = ['Europe','Caribbean','Asia','South America','Central America','Oceania','Middle East','Africa','North America']
# Find the correlation between regions
corr_regions = outbounds.loc[:,regions].corr(method='spearman')
# Create a data frame with most and least correlated region of each region
corr_df = pd.DataFrame(index=regions)
for region in regions:
    correlations = corr_regions[region]
    correlations = correlations.loc[correlations < 1]
    corr_df.loc[region,'Most Correlated'] = correlations.loc[correlations == max(correlations)].index[0]
    corr_df.loc[region,'Least Correlated'] = correlations.loc[correlations == min(correlations)].index[0]
display(corr_df)
# Plot correlation matrix
sns.heatmap(corr_regions, annot=True, annot_kws={'size':'small'}, linewidth=0.2)
plt.show()
```

```python
# General seasonal decomposition
decomposition = sm.tsa.seasonal_decompose(outbounds['Grand Total'])
fig, axes = plt.subplots(3,1, figsize=(5,8), sharex=True)
decomposition.seasonal.plot(ax = axes[0], legend=True, label='Seasonal')
decomposition.trend.plot(ax = axes[1], legend=True, label='Trend')
decomposition.resid.plot(ax = axes[2], legend=True, label='Noise')
plt.show()
```

```python
# Initialize an empty dictionary to store seasonal trend of each region in regions 
regions_seasonal = {}
# Initialize an empty dictionary to store trend of each region in regions
regions_trend = {}
# Get seasonal decomposition of each region in regions
for region in regions:
    region_decomp = sm.tsa.seasonal_decompose(outbounds[region])
    regions_seasonal[region] = region_decomp.seasonal
    regions_trend[region] = region_decomp.trend
# Create a data frame from regions_seasonal dictionary
regions_seasonal_df = pd.DataFrame(regions_seasonal)
# Create a data frame from regions_trend dictionary
regions_trend_df = pd.DataFrame(regions_trend)
```

```python
# Plot seasonal trend
axes = regions_seasonal_df.plot(subplots=True, layout=(3,3), figsize=(20,22))
# Set legend for each subplot
for rows in axes:
    for ax in rows:
        ax.legend(loc='lower right', fontsize='small')
plt.tight_layout()
plt.show()
```

```python
# Plot correlation between each region's seasonality
# Get correlation
seasonality_corr = regions_seasonal_df.corr(method='spearman')
# Plot cluster heatmap
sns.clustermap(seasonality_corr, annot=True, linewidth=0.2)
plt.show()
```

```python
# Plot trend
axes = regions_trend_df.plot(subplots=True, layout=(3,3), figsize=(18,15))
# Set legend for each subplot
for rows in axes:
    for ax in rows:
        ax.legend(loc='lower right', fontsize='medium')
plt.tight_layout()
plt.show()
```

```python
# Test stationary of grand total
df_test = adfuller(outbounds['Grand Total'], autolag='AIC')
df_result = pd.Series(df_test[:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in df_test[4].items():
    df_result['Critical Value (%s)'%key] = value
display(df_result)
```
