```python
# File paths
outbound15 = 'Dataset/15outbound.xls'
outbound16 = 'Dataset/16outbound.xls'
outbound17 = 'Dataset/17outbound.xls'
```

```python
# Create dataframe from excel sheet
def create_df(file_path,sheet,header,cols,skips):
    
    df = pd.read_excel(file_path, sheet_name=sheet, header=header, usecols=cols, nrows=18, skiprows=skips)
    
    return df
```

```python
# 2014 outbound
cols = ['Regions','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
skip_rows = [50, 52, 54, 56, 58]
df14 = create_df(outbound15,0,41,cols,skip_rows)
display(df14)
```

```python
# 2015 outbound
skip_rows = [12, 14, 16, 18, 20]
df15 = create_df(outbound15,0,3,cols,skip_rows)
display(df15)
```

```python
# 2016 outbound
df16 = create_df(outbound16,'Dec 2016',3,cols,skip_rows)
display(df16)
```

```python
# 2017 outbound
cols = ['Regions','Janrev','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
df17 = create_df(outbound17,'December 2017',3,cols,skip_rows)
# Rename January column
df17.rename(columns={'Janrev':'Jan'},inplace=True)
display(df17)
```

```python
# Create datetime index
def to_time(df,year):
    # Set Regions as index
    df.set_index('Regions', inplace=True)
    # Swap index and column
    df = df.stack().unstack(0)
    # Convert months' abbreviation to actual datetime
    months = {df.index[i-1]:str(year)+str(i) for i in range(1,13)}
    df.rename(index=months, inplace=True)
    df.index = pd.to_datetime(df.index, format='%Y%m')+pd.offsets.MonthEnd()
    # Unify columns' name
    df.columns = ['Europe','Caribbean','Asia','South America','Central America','Oceania','Middle East','Africa','Total Overseas','Mexico','Canada','North America','Grand Total']

    return df
```

```python
df14 = to_time(df14,2014)
display(df14)
display(df14.info())
```

```python
df15 = to_time(df15,2015)
display(df15)
display(df15.info())
```

```python
df16 = to_time(df16,2016)
display(df16)
display(df16.info())
```

```python
df17 = to_time(df17, 2017)
display(df17)
display(df17.info())
```

```python
outbounds = pd.concat([df14,df15,df16,df17])
display(outbounds)
```

```python
# Export the finalized dataset as csv file
outbounds.to_csv('Dataset/outbounds14to17.csv', float_format='%.2f', encoding='utf-8')
```
