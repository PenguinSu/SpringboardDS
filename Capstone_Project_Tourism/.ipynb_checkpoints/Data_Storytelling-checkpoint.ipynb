```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from IPython.display import display, HTML

%matplotlib inline 
%load_ext watermark
# Set default style
sns.set()
```

```python
%watermark -v -p pandas,numpy,matplotlib,seaborn -u -n -t -z
```

```python
# File path
outbounds_file = 'Dataset/outbounds14to17.csv'
```

```python
# Read in data as DataFrame
outbounds = pd.read_csv(outbounds_file, index_col=0, parse_dates=True)
display(outbounds)
```

```python
# Overall info
display(outbounds.info())
display(outbounds.describe())
```

```python
# General trend
outbounds['Grand Total'].plot(marker='o')
# Set title and labels
plt.title('Traveling Trend (2014-2017)')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.show()
```

```python
# Each year trend by regions
fig, axes = plt.subplots(2,2, figsize=(10,8))
# Set starting year
year = 2014
# Plot total traveling amount by year
for i in range(0,2):
    for j in range(0,2):
        outbounds.loc[str(year),'Grand Total'].plot(ax=axes[i,j], marker='o')
        # Set title and labels
        axes[i,j].set_title(str(year))
        axes[i,j].set_xlabel('Months')
        axes[i,j].set_ylabel('Amount')
        # Increase year by 1
        year += 1
plt.tight_layout()
plt.show()
```

```python
# Traveling trend by regions
# List of regions subset
regions = ['Europe','Caribbean','Asia','South America','Central America','Oceania','Middle East','Africa','North America']
# Set colors
colors = sns.color_palette('hls',len(regions))
# Plot graph
outbounds.loc[:,regions].plot(marker='o',color=colors, figsize=(10,12))
# Set title and labels
plt.title('Traveling Trend by Regions (2014-2017)',fontsize='large')
plt.xlabel('Months')
plt.ylabel('Amount')
plt.show()
```

```python
# Each year trend by regions
fig, axes = plt.subplots(2,2, figsize=(10,15))
# Set starting year
year = 2014
# Plot graphs
for i in range(0,2):
    for j in range(0,2):
        outbounds.loc[str(year),regions].plot(ax=axes[i,j], marker='o',color=colors)
        # Set title and labels
        axes[i,j].set_title(str(year))
        axes[i,j].set_xlabel('Months')
        axes[i,j].set_ylabel('Amount')
        axes[i,j].legend(loc='upper left', fontsize='x-small')
        # Increase year by 1
        year += 1
plt.tight_layout()
plt.show()
```

```python
# Traveling trend without North America
outbounds.loc[:,'Total Overseas'].plot(marker='o')
# Set title and labels
plt.title('Traveling Trend without North America (2014-2017)')
plt.xlabel('Months')
plt.ylabel('Amount')
plt.show()
```

```python
# Helper function to plot each region by year
def print_years(region, color):
    fig, axes = plt.subplots(2,2, figsize=(13,15))
    # Set starting year
    year = 2014
    # Plot graphs
    for i in range(0,2):
        for j in range(0,2):
            outbounds.loc[str(year),region].plot(ax=axes[i,j], marker='o',color=color)
            # Set title and labels
            axes[i,j].set_title(str(year))
            axes[i,j].set_xlabel('Months')
            axes[i,j].set_ylabel('Amount')
            axes[i,j].legend(loc='upper left', fontsize='x-large')
            # Increase year by 1
            year += 1
    plt.tight_layout()
    plt.show()
```

```python
# Plot regions individually
for r in range(0,len(regions)):
    print_years(regions[r], colors[r])
```

```python
# Check seasonal trend by month
# Get months
index_month = outbounds.index.month
# Get average traveling amount each month
month_mean = outbounds.groupby(index_month, axis='index').mean()
# Plot graphs
month_mean.plot(figsize=(13,16),subplots=True,layout=(7,2),sharey=False)
plt.show()
```

```python
# Resample dataset to quaterly frequency and get the total amount
quaterly_outbounds = outbounds.resample('Q').sum()
display(quaterly_outbounds)
```

```python
# Resample dataset to quaterly frequency and get the average amount
quaterly_mean = outbounds.resample('Q').mean()
display(quaterly_mean)
```

```python
# Set xticks format
yrmonthFmt = quaterly_outbounds.index.strftime('%Y-%m')
# Plot quaterly trend by regions
ax = quaterly_outbounds.loc[:,regions].plot(kind='area',figsize=(13,13), rot='horizontal',stacked=True)
# Seperate each year
ax.axvline('2014-12', color='w', linestyle='--')
ax.axvline('2015-12', color='w', linestyle='--')
ax.axvline('2016-12', color='w', linestyle='--')
# Format x ticks
ax.xaxis.set_major_locator(plt.MaxNLocator(len(quaterly_outbounds.index)))
ax.xaxis.set_major_formatter(plt.FixedFormatter(yrmonthFmt))
# Set title and labels
plt.title('Traveling Trend by Regions (Quaterly)',fontsize='large')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.show()
```

```python
# Plot total traveling amount quaterly with mean
ax = quaterly_outbounds.loc[:,'Grand Total'].plot(figsize=(12,5), marker='o', legend=True)
quaterly_mean.loc[:,'Grand Total'].plot(linestyle='--', legend=True, label='Grand Total (Mean)')
# Format x ticks
ax.xaxis.set_major_locator(plt.MaxNLocator(len(quaterly_outbounds.index)))
ax.xaxis.set_major_formatter(plt.FixedFormatter(yrmonthFmt))
# Set title and labels
plt.title('Traveling Trend with Mean (Quaterly)',fontsize='large')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.show()
```

```python
# Resample dataset to annual frequency and get the total amount
yearly_outbounds = outbounds.resample('A').sum()
display(yearly_outbounds)
```

```python
# Set the xticks format
yearFmt = yearly_outbounds.index.strftime('%Y')
# Plot annual trend by regions
ax = yearly_outbounds.loc[:,regions].plot(kind='area',figsize=(8,15), rot='horizontal',stacked=True)
# Format x ticks
ax.xaxis.set_major_formatter(plt.FixedFormatter(yearFmt))
# Seperate each year
ax.axvline('2015', color='w', linestyle='--')
ax.axvline('2016', color='w', linestyle='--')
# Set title and labels
plt.title('Traveling Trend by Regions (Anually)',fontsize='large')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.show()
```

```python
# Resample dataset to annual frequency and get the average amount
yearly_mean = outbounds.resample('A').mean()
fig, axes = plt.subplots(2,1,figsize=(5,8), sharex=True)
# Plot total and mean
yearly_outbounds.loc[:,'Grand Total'].plot(ax=axes[0], marker='o', legend=True)
yearly_mean.loc[:,'Grand Total'].plot(ax=axes[1], linestyle='--', legend=True, color='orange', label='Grand Total (Mean)')
# Format x ticks
ax.xaxis.set_major_formatter(plt.FixedFormatter(yearFmt))
# Set labels
plt.xlabel('Year')
axes[0].set_ylabel('Amount')
axes[1].set_ylabel('Amount')
plt.show()
```
